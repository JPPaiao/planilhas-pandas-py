```python
import pandas as pd
```

```python
tabela = pd.read_excel('caixa-mensal-2023.xlsx', sheet_name=0)
tabela = tabela.dropna(axis=1, how='all')
tabela = tabela.dropna(axis=0, how='all')
tabela = tabela.drop(["PC", "SEMANA", "NOMES", "MPPC"], axis=1)
display(tabela["DATA"])
# display(tabela)

```

```python
soma_pix = tabela["PIX"].sum()
soma_cartao = tabela["CARTAO"].sum()
soma_ifood = tabela["IFOOD"].sum()
soma_caixa = tabela["CAIXA"].sum()
soma = soma_caixa + soma_cartao + soma_ifood + soma_pix
print(soma_pix, soma_caixa, soma_cartao, soma_ifood)
print(soma)
```

```python
def entrada_ou_saida_mes(colunas):
    total = 0
    soma_entradas = []
    lista_entradas = []

    for col in colunas:
        for contador in range(26):
            total += tabela.at[contador, col]
            contador += 1

        soma_entradas.append(round(total, 2))
        total = 0

    for contador in range(soma_entradas.__len__()):
        lista_entradas.append({
            colunas[contador]: soma_entradas[contador]
        })

    return lista_entradas

print(entrada_ou_saida_mes(tabelas_entrada))
# print(entrada_ou_saida_mes(tabelas_saida))
```

```python
# Total de entrada mensal

c = 0
total = 0
total_entrada_mes = []

for tab in tabelas_entrada:
    for c in range(26):
        total += tabela.at[c, tab]
        c+=1
    total_entrada_mes.append(round(total, 2))
    total = 0

print(total_entrada_mes)
```

```python
def total_dia(colunas, datas):
    linhas = 0
    dias = 1
    total = 0
    total_mensal = 0.00
    entradas_dia = []
    total_diario = []

    for contador in range(26):
        for col in colunas:
            if linhas > 25:
                break
            
            total += tabela.at[linhas, col]

            if col == colunas[-1]:
                dia = {
                    dias: total
                }

                entradas_dia.append(dia)
                total_mensal += total
                
                total = 0
                linhas += 1
                dias += 1

    for contador in range(26):
        total_diario.append({
            datas[contador]: entradas_dia[contador]
        })

    return {
        "total_diario": total_diario,
        "total_mensal": total_mensal
    }

print(total_dia(tabelas_entrada, data))
# print(total_dia(tabelas_saida, data))
```

```python
# Total de caixa diario e mensal

x, z, t = 1, 1, 0
dias = 1
total_mensal = 0.00
total_diario = []

for x in range(26):
    for tab in tabelas_entrada:
        if z > 25:
            break
        
        t += tabela.at[z, tab]

        if tab == "IFOOD":
            dia = {
                dias: round(t, 2)
            }
            total_diario.append(dia)
            
            total_mensal += t
            
            t = 0
            z += 1
            dias += 1
            
print(total_diario)
print(total_mensal)
        
```
