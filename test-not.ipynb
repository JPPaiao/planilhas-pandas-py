```python
import pandas as pd
```

```python
tabela = pd.read_excel('caixa-mensal-2023.xlsx', sheet_name=0)
tabela = tabela.dropna(axis=1, how='all')
tabela = tabela.dropna(axis=0, how='all')
tabela = tabela.drop(["DATA", "PC", "SEMANA", "NOMES"], axis=1)
# display(tabela["TOTAL D"])
display(tabela)

```

```python
soma_pix = tabela["PIX"].sum()
soma_cartao = tabela["CARTAO"].sum()
soma_ifood = tabela["IFOOD"].sum()
soma_caixa = tabela["CAIXA"].sum()
soma = soma_caixa + soma_cartao + soma_ifood + soma_pix
print(soma_pix, soma_caixa, soma_cartao, soma_ifood)
print(soma)
```

```python
print(tabela.at[1, "CAIXA"])
print(tabela.at[1, "PIX"])
print(tabela.at[1, "CARTAO"])
print(tabela.at[1, "IFOOD"])

tabelas_entrada = ["CAIXA", "PIX", "CARTAO", "IFOOD"]
tabelas_saida = ["DIZ", "SALGADO", "EMBALAGENS", "AÇAI/MILK", "FEIRA/PAD.", "COMBUSTIVEL", "MERCADO", "PARTICULAR", "OUTROS"]
```

```python
def entrada_ou_saida_mes(colunas):
    total = 0
    contador = 0
    total_entrada_mes = []

    for col in colunas:
        for contador in range(26):
            total += tabela.at[contador, col]
            contador += 1

        total_entrada_mes.append(round(total, 2))
        total = 0

    return total_entrada_mes

# print(entrada_ou_saida_mes(tabelas_entrada))
print(entrada_ou_saida_mes(tabelas_saida))
```

```python
# Total de entrada mensal

c = 0
total = 0
total_entrada_mes = []

for tab in tabelas_entrada:
    for c in range(26):
        total += tabela.at[c, tab]
        c+=1
    total_entrada_mes.append(round(total, 2))
    total = 0

print(total_entrada_mes)
```

```python
def total_dia(colunas):
    linhas = 1
    dias = 1
    total = 1
    contador = 1
    total_mensal = 0.00
    total_diario = []

    for contador in range(26):
        for col in colunas:
            if linhas > 25:
                break
            
            total += tabela.at[linhas, col]

            if col == colunas[-1]:
                dia = {
                    dias: round(total, 2)
                }

                total_diario.append(dia)
                total_mensal += total
                
                total = 0
                linhas += 1
                dias += 1

    return {
        "total_diario": total_diario,
        "total_mensal": total_mensal
    }

# print(total_dia(tabelas_entrada))
print(total_dia(tabelas_saida))
```
